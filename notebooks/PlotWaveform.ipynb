```python
import os

data_dir = '/home/dfc13/WaveformDeepLearning/data/SPEs/not_spe/'
fname = os.path.join(data_dir, 'D1--not_spe--00002.txt')

#data_dir = '/home/dfc13/DeepLearningWithPython/data/SPEs/d1_shifted/'
#fname = os.path.join(data_dir, 'D1--waveforms--00003.txt')

f = open(fname)
data = f.read()
f.close()

lines = data.split('\n')

# ignore top 5 lines and last line of file
lines = lines[5:len(lines)-1]

#print(header)
#print(len(lines))
```

```python
import numpy as np

times = np.zeros(len(lines),dtype='float')
ADCs  = np.zeros(len(lines),dtype='float')
for i,line in enumerate(lines):
    if line:
        times[i] = float(line.split(',')[0])*1.e9
        ADCs[i] = float(line.split(',')[1])
```

```python
from matplotlib import pyplot as plt

plt.plot(times,ADCs)
plt.show()
```
